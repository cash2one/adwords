```python
import csv
import codecs

file_name = 'excel.csv'

basic_dict = {} #{destination id : destination name}

ready_dict = {} #{destinaiton name: [destination slug, destination id, destination type, number of listings, url, country name]}

#ready_dict['key'][0] = destination id
#ready_dict['key'][1] = destination type
#ready_dict['key'][2] = number of listings
#ready_dict['key'][3] = URL (landing page)
#ready_dict['key'][4] = country name

ready_urls = []
ready_names = []

#set up basic_dict, ready_urls and ready_names
with codecs.open(file_name, 'r', encoding = 'utf-8', errors = 'ignore') as csvfile:
    reader=csv.DictReader(csvfile)
    for row in reader:
        ready_urls.append(row['concat']) #add all urls
        if '_' in row['name']: #if the name of the destination is messed up (e.g. Br_zo)
            basic_dict[int(row['destination_id'])] = row['slug1'].replace("-", " ").title() #convert slug as the name
            ready_names.append(row['slug1'].replace("-", " ").title()) #add all the corrected names of the destination
        else: #if the name of the destination is fine
            basic_dict[int(row['destination_id'])] = row['name'] 
            ready_names.append(row['name']) #add all the alerady-correct names of the destination
            
        
        
for i in range(len(ready_urls)):
    
```
